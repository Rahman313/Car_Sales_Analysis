```python
genderwise_buyer=df.groupby('Gender')['Transmission'].value_counts().unstack()
genderwise_buyer=genderwise_buyer.plot(kind='bar')
plt.title("Gender wise Transmisson Buyer of car")
plt.xlabel("Gender")
plt.ylabel("Transmission")
plt.xticks(rotation=0)
plt.legend(['Auto','Manual'])
plt.show()
```

```python
plt.pie(df['Dealer_Region'].value_counts(),labels=df['Dealer_Region'].value_counts().index,autopct='%1.2f%%')
plt.legend(loc='upper right',bbox_to_anchor=(0, 1.3))
plt.show()
```

```python
plt.pie(df['Body Style'].value_counts(),labels=df['Body Style'].value_counts().index,shadow='True',autopct='%1.2f%%')
plt.legend(loc='upper right',bbox_to_anchor=(0, 1.3))
plt.show()
```

```python
ex=[0.1,0.1,0.0]
plt.pie(df['Color'].value_counts(),labels=df['Color'].value_counts().index,shadow='True',explode=ex,autopct='%1.2f%%')
plt.legend(loc='upper right',bbox_to_anchor=(1,1.3))
plt.show()
```
